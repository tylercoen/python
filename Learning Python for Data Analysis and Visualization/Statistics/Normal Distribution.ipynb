```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from scipy import stats

# set the mean
mean = 0

# set the standard deviation
std = 1

# create a range
X = np.arange(-4,4,0.01)

# create the normal distribution for the range
Y = stats.norm.pdf(X,mean,std)

plt.plot(X,Y)
x_fill = np.linspace(-1,1,1)
y_fill = stats.norm.pdf(x_fill,mean,std)
plt.fill_between(x_fill,y_fill,0)
```

```python
# using numpy to create the normal distribution

# set the mean and std
mu,sigma = 0,0.1

# now grab 1000 random numbers from the normal distribution

norm_set = np.random.normal(mu,sigma,1000)
```

```python
# plot using seaborn
import seaborn as sns

plt.hist(norm_set,bins=50)
```
