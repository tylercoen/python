```python
mean_obama = poll_df["Obama"].mean()
mean_romney = poll_df['Romney'].mean()
mean_undecided = poll_df['Undecided'].mean()
```

```python
avg = DataFrame({'Obama':[mean_obama],'Romney':[mean_romney],
                'Undecided':[mean_undecided]})
```

```python
avg
```

```python
std_obama = poll_df["Obama"].std()
std_romney = poll_df['Romney'].std()
std_undecided = poll_df['Undecided'].std()



```

```python
std = DataFrame({'Obama':std_obama,'Romney':std_romney,
                 'Undecided':std_undecided},index=[0])
```

```python
std
```

```python
avg.plot(yerr=std,kind='bar',legend=False)
```

```python
poll_avg = pd.concat([avg,std],axis=0)
```

```python
poll_avg = poll_avg.transpose()
```

```python
poll_avg.columns = ['Average','STD']
```

```python
poll_avg
```

```python
party_map = {'Bachmann, Michelle': 'Republican', 'Romney, Mitt': 'Republican', 'Obama, Barack': 'Democrat',
       "Roemer, Charles E. 'Buddy' III": 'Republican', 'Pawlenty, Timothy': 'Republican',
       'Johnson, Gary Earl': 'Republican', 'Paul, Ron': 'Republican', 'Santorum, Rick': 'Republican',
       'Cain, Herman': 'Republican', 'Gingrich, Newt': 'Republican', 'McCotter, Thaddeus G': 'Republican',
       'Huntsman, Jon': 'Republican', 'Perry, Rick': 'Republican'}
```

```python
donor_df['Party'] = donor_df.cand_nm.map(party_map)
```

```python
donor_df = donor_df[donor_df.contb_receipt_amt > 0]
```

```python
occupation_df = donor_df.pivot_table('contb_receipt_amt',
                                    index='contbr_occupation',
                                    columns = 'Party',
                                    aggfunc='sum')
```

```python
occupation_df.tail()
```

```python
occupation_df.shape
```

```python
occupation_df = occupation_df[occupation_df.sum(1) > 1000000]
```

```python
occupation_df.shape
```

```python
occupation_df.plot(kind='bar')
```

```python
occupation_df.plot(kind='barh',figsize=(10,12),cmap='seismic')
```

```python
occupation_df.drop(['INFORMATION REQUESTED PER BEST EFFORTS',
                    'INFORMATION REQUESTED'],axis=0,inplace=True)
```

```python
occupation_df.loc['CEO'] = occupation_df.loc['CEO'] + occupation_df.loc['C.E.O.']

occupation_df.drop('C.E.O.',inplace=True)
```
