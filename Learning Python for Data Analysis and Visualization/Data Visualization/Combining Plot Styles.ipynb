```python
# the normal imports
import numpy as np
from numpy.random import randn

# import the stats library from numpy
from scipy import stats

# plotting modules and libraries:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# make plots visible in jupyter
%matplotlib inline
```

```python
dataset = randn(100)
```

```python
sns.distplot(dataset,bins=25)

"""
`distplot` is a deprecated function and will be removed in seaborn v0.14.0.

Please adapt your code to use either `displot` (a figure-level function with
similar flexibility) or `histplot` (an axes-level function for histograms).

"""
```

```python
sns.distplot(dataset,bins=25,rug=True,hist=False)
```

```python
sns.histplot(data=dataset,bins=25)
```

```python
sns.displot(data=dataset,bins=25)
```

```python
sns.distplot(dataset,bins=25,kde_kws={'color':'indianred','label':'KDE PLOT'},
            hist_kws={'color':'blue','label':'HIST'})
```
