```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# There are the plotting modules and libraries:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the notebook
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,order=4,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False)
```

```python
tips.head()
```

```python
tips['tips_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot(x='size',y='tips_pect',data=tips)
```

```python
sns.lmplot(x='size',y='tips_pect',data=tips,x_jitter=.1)
```

```python
sns.lmplot(x='size',y='tips_pect',data=tips,x_estimator=np.mean)
```

```python
sns.lmplot(x='total_bill',y='tips_pect',data=tips,hue='sex',
           markers=['x','o'])
```

```python
sns.lmplot(x='total_bill',y='tips_pect',data=tips,hue='day')
```

```python
sns.lmplot(x='total_bill',y='tips_pect',data=tips,lowess=True,
          line_kws={'color':'black'})
```

```python
sns.regplot(x='total_bill',y='tips_pect',data=tips)
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)

sns.regplot(x='total_bill',y='tips_pect',data=tips,ax=axis1)
sns.violinplot(x=tips['tips_pect'],y=tips['size'],data=tips,color='Reds_r',
               ax=axis2)
```
