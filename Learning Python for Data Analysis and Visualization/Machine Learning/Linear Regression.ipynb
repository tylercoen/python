```python
from sklearn.datasets import load_boston
```

```python
boston = pd.read_csv('Boston-house-price-data.csv')
```

```python
print(boston.describe())
```

```python
boston.head()
```

```python
plt.hist(boston.MEDV,bins=50)
# MEDV is the target
plt.xlabel('Prices in $10,000s')
plt.ylabel('Number of houses')
```

```python
plt.scatter(boston.RM,boston.MEDV)

plt.ylabel('Price in $10,000s')
plt.xlabel('Number of rooms')
```

```python
boston_df = DataFrame(boston)
```

```python
boston_df.head()
```

```python
boston_df['Price'] = boston.MEDV
```

```python
boston_df.head()
```

```python
sns.lmplot(x='RM',y='Price',data=boston_df)
```

```python
X = boston_df.RM
```

```python
X
```

```python
X =np.vstack(boston_df.RM)
```

```python
X.shape
```

```python
Y = boston_df.Price
```

```python
Y
```

```python
# Create the X array in the form [X 1]
X = np.array([[value[0],1] for value in X])
```

```python
X
```

```python
m,b = np.linalg.lstsq(X,Y,rcond=None)[0]
```

```python
plt.plot(boston_df.RM,boston_df.Price,'o')

x = boston_df.RM
plt.plot(x, m*x + b,'r',label='Best Fit Line')
```

```python
result = np.linalg.lstsq(X,Y,rcond=None)

error_total = result[1]

rmse = np.sqrt(error_total/len(X))

print('The root mean square error was %.2f' %rmse)
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()
```

```python
X_multi = boston_df.drop(columns=['Price','MEDV'])
Y_target = boston_df.Price
```

```python
lreg.fit(X_multi,Y_target)
lreg
```

```python
print('The estimated intercept coeffecient is %.2f' %lreg.intercept_)
print('The number of coeffectiences used was %d' %len(lreg.coef_))
```

```python
coeff_df = DataFrame(boston_df.columns)
coeff_df.columns = ['Features']
coeff_df['Coefficient Estimate'] = Series(lreg.coef_)

coeff_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, boston_df['Price'], test_size=0.2, random_state=42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
lreg = LinearRegression()

lreg.fit(X_train,Y_train)
```

```python
pred_train = lreg.predict(X_train)
pred_test = lreg.predict(X_test)
```

```python
print('Fit a model X_train, and calculate the MSE with Y_train: %.2f' %np.mean((Y_train-pred_train)**2))
print('Fit a model X_train and calculate the MSE with X_test and Y_test: %.2f' %np.mean((Y_test-pred_test)**2))
```

```python
train = plt.scatter(pred_train,(pred_train- Y_train),c='b',alpha=0.5)

test = plt.scatter(pred_test,(pred_test-Y_test),c='r',alpha=0.5)

plt.hlines(y=0,xmin=-10,xmax=40)

plt.legend((train,test),('Training','Test'),loc='lower left')

plt.title('Residual Plots')
```
