** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
df1 = df[df['JobTitle'].str.contains('Chief',case=False)]
df1
```

```python
indices_to_drop = []

for index, row in df1.iterrows():
    if 'Chief,' in row['JobTitle']:
        indices_to_drop.append(index)

df1_copy = df1.copy()  # Create a copy of the DataFrame
df1_copy.drop(indices_to_drop, inplace=True)
df1_copy
```
